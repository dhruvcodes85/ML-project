```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv("Algerian_forest_fires_dataset_UPDATE.csv", header=1) # header tells us which row to consider for features
dataset
```

```python
dataset.info()
```

## Data cleaning

```python
## missing values

dataset.isnull().sum()
```

```python
dataset[dataset.isnull().any(axis=1)]
```

The dataset is converted into two sets based on Region from 122th index, we can make a new
column based on the Region

1 : "Bejaia Region Dataset"

2 : "Sidi-Bel Abbes Region Dataset"

Add new column with region

```python
## We can see that the second region's heading is also considered as a row value so lets remove it seeing its index

dataset.loc[:122, 'Region'] = 0
dataset.loc[122:, 'Region'] = 1
dataset[['Region']] = dataset[['Region']].astype(int)
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
## removing the null values

# df = dataset.dropna().reset_index(drop=False) # by default drop = False which means it preserves the old index values in a new col named index it is unwanted
df = dataset.dropna().reset_index(drop=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
## Removing the 122nd row

df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## Cols consists of extra spaces, so let's remove it

df.columns = df.columns.str.strip()
df.columns
```

```python
 df.info()
```

## Changing the required cols to int datatype

```python
df.head()
```

```python
df.columns
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
df.info()
```

```python
# df[['Rain', 'FFMC', 'DMC', 'DC', 'ISI', 'BUI', 'FWI']] = df[['Rain', 'FFMC', 'DMC', 'DC', 'ISI', 'BUI', 'FWI']].astype(float)
```

```python
df.info()
```

## Converting all other features to float dtype but using an efficient way

```python
objects = [features for features in df.columns if df[features].dtype == 'O']
objects
```

```python
for i in objects:
    if i != 'Classes':
        df[i] = df[i].astype(float)
df.head()
```

```python
df.info()
```

```python
## can even save the cleaned dataset 
df.to_csv('Algerian_forest_fires_cleaned_dataset_practically.csv', index = False) # index=False indicated that I don't want to save my indices in a new column
```
